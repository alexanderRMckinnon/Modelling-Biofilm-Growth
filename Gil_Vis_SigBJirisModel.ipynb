Phosphoralation of V:
- Low stress state: $\ce{V <=>> \underline{\mathbf{V^{p}}}}$. High stress state: $\ce{\underline{\mathbf{V}} <<=> V^{p}}$

$V$ Binding with $W$:
- Low stress state: $\ce{\underline{\mathbf{W}} + V <<=> W:V}$. High stress state: $\ce{W + V <=>> \underline{\mathbf{W:V}}}$

$\sigma^B$ Binding with $W$:
- Low stress state: $\ce{\sigma^{B} + W <=>> \underline{\mathbf{\sigma^{B}:W}}}$. High stress state: $\ce{\underline{\mathbf{\sigma^{B}}} + W <<=>\sigma^{B}:W}$

Self Induction:
- $\ce{\sigma^{B} -> \sigma^{B} + V + W}$

Now we can re-write this more biologically accurate model as propensity functions:
 - $Reaction, \qquad Propensity$
 
Production of $\sigma^B, V, W$:
 - $\ce{SigB -> SigB + 1}, \qquad \gamma \times SigB + \epsilon$
 - $\ce{V -> V + 1}, \qquad \gamma \times SigB + \epsilon$
 - $\ce{W -> W + 1}, \qquad \gamma \times SigB + \epsilon$
 
Phosphoralation of V:
 - $\ce{V -> V - 1}, \qquad \alpha_p \times V$
 - $\ce{Vp -> Vp + 1}, \qquad \alpha_p \times V$

Dephosphoralation of V:
 - $\ce{V -> V + 1}, \qquad \beta_p \times V_p \times \xi$
 - $\ce{Vp -> Vp - 1}, \qquad \beta_p \times V_p \times \xi$
 
$V$ binding with $W$:
 - $\ce{W -> W - 1}, \qquad \alpha_{vw} \times V \times W$
 - $\ce{V -> V - 1}, \qquad \alpha_{vw} \times V \times W$
 - $\ce{VW -> VW + 1}, \qquad \alpha_{vw} \times V \times W$
 
$V$ unbinding with $W$:
 - $\ce{W -> W + 1}, \qquad \beta_{vw} \times VW$
 - $\ce{V -> V + 1}, \qquad \beta_{vw} \times VW$
 - $\ce{VW -> VW - 1}, \qquad \beta_{vw} \times VW$

$\sigma^B$ binding with $W$:
 - $\ce{SigB -> SigB - 1}, \qquad \alpha_{SigBW} \times SigB \times W$
 - $\ce{W -> W - 1}, \qquad \alpha_{SigBW} \times SigB \times W$
 - $\ce{SigBW -> SigBW + 1}, \qquad \alpha_{SigBW} \times SigB \times W$
 
$\sigma^B$ unbinding with $W$:
 - $\ce{SigB -> SigB + 1}, \qquad \beta_{SigBW} \times SigBW$
 - $\ce{W -> W + 1}, \qquad \beta_{SigBW} \times SigBW$
 - $\ce{SigBW -> SigBW - 1}, \qquad \beta_{SigBW} \times SigBW$
 
Free species degradation (not simultaneous):
 - $\ce{SigB -> SigB - 1}, \qquad \delta_f \times SigB$
 - $\ce{V -> V - 1}, \qquad \delta_f \times V$
 - $\ce{W -> W - 1}, \qquad \delta_f \times W$
 
Bound species degradation (not simultaneous):
 - $\ce{Vp -> Vp - 1}, \qquad \delta_b \times Vp$
 - $\ce{VW -> VW - 1}, \qquad \delta_b \times VW$
 - $\ce{SigBW -> SigBW - 1}, \qquad \delta_b \times SigBW$

To gain initial parameter estimation, we will use paper [Quantitative Aspect of Bacillus subtilis σB Regulatory Network—A Computational Simulation](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC9775250/). This paper has a much more complex model, which is described below:
1. Binding of anti-sigma factor $W_2$ and anti-anti-sigma factor $V$:
    - $\ce{W_{2} + V <=>[k_{1+}][k_{1-}] W_{2}V}, \qquad \ce{W_{2}V + V <=>[k_{2+}][k_{2-}] W_{2}V_{2}}$
2. Phosphorylation of anti-anti-sigma factor $V$ by anti-simga factor $W_2$:
    - $\ce{W_{2} + V + p ->[k_{3+}] W_{2} + Vp}, \qquad \ce{W_{2}V_{2} + p ->[k_{4+}] W_{2}V + Vp}$
3. Binding of sigma factor $\sigma^B$ to anti-sigma factor $W_2$:
    - $\ce{W_{2} + \sigma^{B} <=>[k_{5+}][k_{5-}] W_{2}\sigma^{B}}$
4. Displacement of sigma factor $\sigma^B$ by anti-anti-sigma factor $V$ from anti-sigma factor $W_2$:
    - $\ce{W_{2}\sigma^{B} + V <=>[k_{6+}][k_{6-}] W_{2}V + \sigma^{B}}$
5. Dephosphorylation of anti-anti-sigma factor $VP$ by stress signals $U$ and $P$:
    - $\ce{Vp + U ->[k_{7+}] V + p + U}, \qquad \ce{Vp + P ->[k_{7+}] V + p + P}$

```python
import multiprocessing
from tqdm.notebook import tqdm
import numpy as np
import scipy.stats as st
import numba

# Plotting modules
import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()

# Line profiler (can install with conda install line_profiler)
%load_ext line_profiler

def sample_discrete(probabilities):
    # Generate random number
    r_num = np.random.rand()
    
    # Find assosiated propensity
    ind = 0
    prob_sum = 0.0
    while prob_sum < r_num:
        prob_sum += probabilities[ind]
        ind += 1
    return ind - 1

def gillespie_draw(propensity_func, propensities, variables, t, args=()):
    # 1) Calculate new propensites
    propensity_func(propensities, variables, t, *args)
    
    # 2) Sum propensities
    propensity_sum = propensities.sum()
    
    # 3) Compute next time
    dt = np.random.exponential(1.0 / propensity_sum)
    
    # 4) Compute reaction probabilites from propensities
    reaction_probabilites = propensities / propensity_sum
    
    # 5) Draw a random reaction
    chosen_reaction = sample_discrete(reaction_probabilites)
    
    return chosen_reaction, dt

# Now we will construct a functions to run the gilespie algorithim
def gillespie_algorithim(propensity_function, update_array, initial_conditions, time_points, args=()):
    # Initialise output
    pop_out = np.empty((len(time_points), update_array.shape[1]), dtype = int)

    # Initialise and perform simulation
    index = 0
    time_index = 1
    time = time_points[0]
    variables = initial_conditions.copy()
    pop_out[0,:] = variables
    propensities = np.zeros(update_array.shape[0])
    while index < len(time_points):
        while time < time_points[time_index]:
            # Draw reaction and time step
            reaction, dt = gillespie_draw(propensity_function, propensities, variables, time, args)

            # Update variables
            variables_previous = variables.copy()
            variables += update_array[reaction, :]

            # Increment time
            time += dt

        # Update index
        index = np.searchsorted(time_points > time, True)

        # Update the variables
        pop_out[time_index:min(index, len(time_points))] = variables_previous

        # Increment index
        time_index = index

    return pop_out
```

# Torkel's SigB Model
Looking at Torkel's paper "The σB alternative sigma factor circuit modulates noise to generate different types of pulsing dynamics".

## Outline of the model:
Three species sigma factor $\sigma^B$, anti-sigma factor $W$, anti-anti-sigma factor $V$. In this model they are each produced at a hill function rate to model the transcription and translation step.
$$\require{mhchem}$$
$$\ce{\emptyset ->[v_{0}\frac{1 + F(\sigma^B)}{K+(\sigma^{B})}] \sigma^{B}}$$
$$\ce{\emptyset ->[\lambda_{W}v_{0}\frac{1 + F(\sigma^B)}{K+(\sigma^{B})}] V}$$
$$\ce{\emptyset ->[\lambda_{W}v_{0}\frac{1 + F(\sigma^B)}{K+(\sigma^{B})}] W}$$

The anti-sigma factor $W$ must first dimerise:
$$\ce{2W <=>[k_{Bw}][k_{Dw}] W_{2}}$$

The anti-sigma factor $W$ can bind to the anti-anti sigma factor $V$ as well as sigma factor $V$:
$$\ce{W_{2} + V <=>[k_{B1}][k_{D1}] W_{2} V}$$

$$\ce{W_{2} V + V <=>[k_{B2}][k_{D2}] W_{2} V_{2}}$$

$$\ce{W_{2} + \sigma^{B} <=>[k_{B3}][k_{D3}] W_{2} \sigma^{B}}$$

$$\ce{P + V^{P} <=>[k_{B4}][k_{D4}] P V^{P}}$$
